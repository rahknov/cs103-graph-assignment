```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_data = pd.read_csv("./datasets/train.csv")
smoking_trend =pd.read_csv("./datasets/smoking.csv")
icecream_sales = pd.read_csv("./datasets/icecream sales.csv")

training_data.head(12)

```

```python
barplot_survival = sns.countplot(x='Survived',data=training_data, hue="Sex")
barplot_survival.set(xlabel= "Sex", ylabel="Number of Persons", title="Survival Frequency")
```

```python
labels=["Class 1", "Class 2", "Class 3"]
# pie_plot = plt.pie(training_data["Pclass"],labels=labels)
parsed_data = training_data["Pclass"].value_counts()

plt.pie(parsed_data, labels=labels,autopct="%1.1f%%")
sns.set_style("darkgrid")
plt.title("Classes of Passengers")
```

```python
line_graph = sns.lineplot(x="Year", y="Data.Percentage.Total", data=smoking_trend)
line_graph.set(ylabel = "Smoking Percentage Total", title="Smoking Percentage From 1980")
```

```python
scatter_plot = sns.scatterplot(x='Temperature', y='Ice Cream Profits', data=icecream_sales, color="red")
icecream_sales.head(15)
scatter_plot.set(title = "Ice Cream Sales and Temperature Relationship")
```

```python
histogram = sns.histplot(data=training_data, x="Age",color="Aquamarine")
plt.title("Age Distributions of Titanic Passengers")
```
